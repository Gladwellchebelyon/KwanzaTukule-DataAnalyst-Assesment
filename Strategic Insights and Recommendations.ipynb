## 2. Customer Retention

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# Read the cleaned data
df = pd.read_csv('transaction_data_cleaned_and_engineered.csv')

def analyze_customer_retention(df):
    # Calculate metrics per business per month
    business_monthly = df.groupby(['ANONYMIZED BUSINESS', 'Month-Year']).agg({
        'VALUE': ['count', 'sum', 'mean'],
        'QUANTITY': 'sum'
    }).reset_index()
    
    # Pivot for frequency analysis
    freq_matrix = business_monthly.pivot(
        index='ANONYMIZED BUSINESS',
        columns='Month-Year',
        values=('VALUE', 'count')
    ).fillna(0)
    
    # Calculate decline indicators
    recent_months = freq_matrix.columns[-3:]
    historic_months = freq_matrix.columns[-6:-3]
    
    recent_avg = freq_matrix[recent_months].mean(axis=1)
    historic_avg = freq_matrix[historic_months].mean(axis=1)
    
    # Identify declining customers
    decline_threshold = 0.5
    declining_customers = freq_matrix[
        (recent_avg < historic_avg * (1-decline_threshold))
    ].index
    
    print(f"\nIdentified {len(declining_customers)} declining customers")
    
    return declining_customers, business_monthly

# Execute analysis
declining_customers, monthly_data = analyze_customer_retention(df)


# Plot retention metrics
plt.figure(figsize=(12, 6))
monthly_active = monthly_data.groupby('Month-Year')['ANONYMIZED BUSINESS'].nunique()
plt.plot(monthly_active.index, monthly_active.values, marker='o')
plt.title('Monthly Active Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
